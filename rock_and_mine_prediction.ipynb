```python
# Import libraries
from IPython import get_ipython
from IPython.display import display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
import warnings
warnings.filterwarnings("ignore")

# Create model list
models=[]
models.append(("lg",LogisticRegression()))
models.append(("dt",DecisionTreeClassifier()))
models.append(("rf",RandomForestClassifier()))
models.append(("knn",KNeighborsClassifier()))
models.append(("svm",SVC(gamma="auto")))
models.append(("nb",GaussianNB()))

# Load dataset
data=pd.read_csv("/content/Copy of sonar data.csv",header=None)
data.head()

# m=mine and r=rock

# Get the values counts of column 60
data.value_counts(60)

# Group by column 60 and get the mean of each group
data.groupby(60).mean()

# Prepare the features and target
x=data.drop(columns=60,axis=1)
y=data[60].values.reshape(-1,1) # changed code. First we need to call .values and then call reshape
# y=data[60].reshape(-1,1) # this line is commented out because it was the root cause of the error.

x
y

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,stratify=y,random_state=1) # Fix the error where the order of variables are incorrect

# Cross Validation
result=[]
name=[]

for name,model in models:
  kfold=StratifiedKFold(n_splits=10,random_state=1,shuffle=True)
  cv_result=cross_val_score(model,x,y,cv=kfold,scoring="accuracy")
  result.append(name)
  print("cv_result",cv_result)
  print("%s: %f (%f)"%(name,cv_result.mean(),cv_result.std()))
  print("")

# Train a Logistic regression model
lg = LogisticRegression()
lg.fit(x_train, y_train)
```

```python
#accuracy in train data
train_pred=lg.predict(x_train)
train_accuracy=accuracy_score(y_train,train_pred)
print("train_accuracy",train_accuracy)
```

```python
#test accuracy
test_pred=lg.predict(x_test)
test_accuracy=accuracy_score(y_test,test_pred)
print("test_accuracy",test_accuracy)
```
